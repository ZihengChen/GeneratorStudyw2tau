```python
from pylab import *
import pandas as pd


#%qtconsole --style monokai
%matplotlib inline
plt.style.use('classic')
plt.rc("figure",facecolor="w",figsize=(6,4))
plt.rc("font",size=10)
plt.rc("savefig",dpi=300)
```

# 1. ttbar_mutau_mumu

```python
df  = pd.read_csv("data/ttbar_mutau_mumu.csv")
df1 = df[df.lepton1_charge==1 ] #direct-leading/ indirect-trailing
df2 = df[df.lepton1_charge==-1] #indirect-leading/ direct-trailing
```

```python
figure = plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
_ = plt.hist([df2.lepton1_pt,df1.lepton1_pt],
         bins=np.arange(0,150,5),alpha = 0.6,
         stacked=True,histtype="stepfilled",lw=0.5,
         color=['deepskyblue','lightskyblue'],normed=True,
         label=['Indirect mu','Direct mu'])
         #label=[r'$\mu$ in $t\rightarrow\tau\rightarrow\mu$',r'$\mu$ in $t\rightarrow\mu$'])   
plt.grid(True)
plt.legend()
#plt.ylim(0,0.03)
plt.xlabel("mu1_pT [GeV]")


plt.subplot(1,2,2)
_ = plt.hist([df2.lepton2_pt,df1.lepton2_pt],
         bins=np.arange(0,150,5),alpha = 0.6,
         stacked=True,histtype="stepfilled",lw=0.5,
         color=['lightskyblue','deepskyblue'],normed=True,
         label=['Direct mu','Indirect mu'])
         #label=[r'$\mu$ in $t\rightarrow\mu$',r'$\mu$ in $t\rightarrow\tau\rightarrow\mu$'])   
plt.grid(True)
plt.ylim(0,0.040)
plt.legend()
plt.xlabel("mu2_pT [GeV]")


plt.savefig("plot/mutau_mumu_pT_stacked")
```

```python
figure = plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
_ = plt.hist([df2.lepton1_pt,df1.lepton1_pt],
         bins=np.arange(0,150,5),alpha = 1,
         stacked=False,histtype="step",lw=2,
         color=['deepskyblue','lightskyblue'],normed=True,
         label=['Indirect mu','Direct mu'])
         #label=[r'$\mu$ in $t\rightarrow\tau\rightarrow\mu$',r'$\mu$ in $t\rightarrow\mu$'])   
plt.grid(True)
plt.legend()
#plt.ylim(0,0.03)
plt.xlabel("mu1_pT [GeV]")


plt.subplot(1,2,2)
_ = plt.hist([df2.lepton2_pt,df1.lepton2_pt],
         bins=np.arange(0,150,5),alpha = 1,
         stacked=False,histtype="step",lw=2,
         color=['lightskyblue','deepskyblue'],normed=True,
         label=['Direct mu','Indirect mu'])
         #label=[r'$\mu$ in $t\rightarrow\mu$',r'$\mu$ in $t\rightarrow\tau\rightarrow\mu$'])   
plt.grid(True)
plt.ylim(0,0.050)
plt.legend()
plt.xlabel("mu2_pT [GeV]")


plt.savefig("plot/mutau_mumu_pT_normed")
```

```python
leptonDirectpt   = np.r_[df[df.lepton1_charge==1].lepton1_pt,df[df.lepton2_charge==1].lepton2_pt]
leptonIndirectpt = np.r_[df[df.lepton1_charge==-1].lepton1_pt,df[df.lepton2_charge==-1].lepton2_pt]
```

```python
_ = hist(leptonDirectpt,
     bins=np.arange(0,150,5),histtype="step",lw=2,
     color='lightskyblue',label='Direct mu',normed=True)
_ = hist(leptonIndirectpt,
     bins=np.arange(0,150,5),histtype="step",lw=3,
     color='deepskyblue',label='Indirect mu',normed=True)
plt.grid(True)
plt.legend()
plt.xlabel("mu_pT [GeV]")
plt.savefig("plot/mutau_mumu_pT_dis")
```

```python
figure = plt.figure(figsize=(5,4))
mutau = plt.hist2d(df.lepton1_pt,df.lepton2_pt,
                   bins=(np.arange(1,100,2),np.arange(1,100,2)),
                   cmap="GnBu",normed=True,vmax=0.0009
                  )
plt.colorbar()
plt.grid(True,c="k")
plt.text(10,85, r"$t\bar{t} \rightarrow \mu\tau_{\mu} \rightarrow \mu\mu$",fontsize=18)
plt.xlabel("lepton1_pT [GeV]")
plt.ylabel("lepton2_pT [GeV]")
plt.savefig("plot/mutau_mumu_pT2D")
```
