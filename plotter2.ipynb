```python
from utility_DNN import *
net = torch.load("checkpointDNN.pt")
net = net.cpu()

removecollist = ['lepton1_charge','lepton1_mother','lepton2_charge','lepton2_mother',
                 'lepton1_phi','lepton2_phi','dilepton_phi',
                 'jet1_phi',"jet2_phi",'dijet_phi','lb1_phi','lb2_phi','met','mpt','metphi'
                ]
#df0nn = df0.drop(removecollist, axis=1)
#df1nn = df1r.drop(removecollist, axis=1)
#df2nn = df2r.drop(removecollist, axis=1)
```

```python
for idf in [df0r,df1r,df2r]:
    temp = idf.drop(removecollist, axis=1)
    temp['nntag'] = np.ones(len(temp.index)).astype(int)
    tempset    = MyDataset(temp.as_matrix())
    
    temploader = DataLoader(tempset, batch_size=tempset.__len__(),shuffle=False, num_workers=1)
    tempiter = iter(temploader).next()
    tempinputs   = tempiter["feature"]
    templabels   = tempiter["label"]
    tempoutputs  = net(Variable(tempinputs))
    temppredicts = torch.max(tempoutputs.data, 1)[1] # return likelihood, predict
    idf["nntag"] = temppredicts.numpy()
```
